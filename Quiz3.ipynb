```python
def get_cosine_similarity(vector1: list[float], vector2: list[float]) -> float:
    return np.dot(vector1, vector2)
```

```python
# methods for file1
methods_file1 = [
"""
def add(a, b):
return a + b
""",
"""
def sum_numbers(x, y):
return x + y
""",
"""
def multiply(a, b):
return a * b
""",
"""
def factorial(n):
if n == 0:
return 1
return n * factorial(n-1)
""",
"""
def contains_substring(text, substring):
return substring in text
"""
]

# methods for file2
methods_file2 = [
"""
def to_uppercase(text):
return text.upper()
""",
"""
def to_lowercase(text):
return text.lower()
""",
"""
def reverse_string(text):
return text[::-1]
""",
"""
def is_palindrome(text):
cleaned = text.replace(" ", "").lower()
return cleaned == cleaned[::-1]
""",
"""
def count_words(text):
return len(text.split())
"""
]
```

```python
def save_to_py_file(filename, methods_list):
    with open(filename, "w") as f:

        content = "\n".join(methods_list)
        f.write(content)
    print(f"File '{filename}' has been created.")

save_to_py_file("file1.py", methods_file1)
save_to_py_file("file2.py", methods_file2)
```

```python
response1 = co.embed(texts=methods_file1, model='embed-english-v3.0', input_type='search_document')
response2 = co.embed(texts=methods_file2, model='embed-english-v3.0', input_type='search_document')
```

```python
e1 = response1.embeddings
e2 = response2.embeddings
```

```python
for i, current_vec in enumerate(e1):

    sim_to_own = np.mean([get_cosine_similarity(current_vec, other) for j, other in enumerate(e1) if i != j])


    sim_to_other = np.mean([get_cosine_similarity(current_vec, other) for other in e2])

    if sim_to_other > sim_to_own:
        print(f"Candidate to move from File 1 to File 2:\n{methods_file1[i]}\n")


for i, current_vec in enumerate(e2):

    sim_to_own = np.mean([get_cosine_similarity(current_vec, other) for j, other in enumerate(e2) if i != j])


    sim_to_other = np.mean([get_cosine_similarity(current_vec, other) for other in e1])

    if sim_to_other > sim_to_own:
        print(f"Candidate to move from File 2 to File 1:\n{methods_file2[i]}\n")
```
